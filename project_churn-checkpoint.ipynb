```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("customer_churn.csv")  # df=== dataframe in pandas 
df.head(5)  # df.head() === used for showing starting 5 data 
```

```python
df.info() # df.info == show the type of data is there  ,,, and object is ===alphanumeric values
```

```python
df["TotalCharges"]=df["TotalCharges"].replace(" ","0") # here we are converting blank value into 0
df["TotalCharges"]=df["TotalCharges"].astype("float")  # now we are converting  int values to  float values
```

```python
# HERE WE ARE CONVERTING THE SENIOR CITIZEN  INT DATATYPE INTO  OBJECT

def convert(values):
    if values == 1:
        return "Yes"
    else:
        return "NO"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(convert)
```

```python
df.head()
```

```python
# HERE WE WILL FIND HOW MANY COUSTOMER STOP USEING CHURN
ax=sns.countplot(x= "Churn" ,data =df)
ax.bar_label(ax.containers[0])
plt.title(" COUNT OF CUSTOMER BY CHURN ")   #  HERE WE WILL GIVE THE  TITLE
plt.show()
```

```python
# how many  customer chun out 
# showing data numericaly
gb=df.groupby('Churn').agg({'Churn':'count'})
print(gb)

# showing data by pie chart
plt.figure(figsize =(3,4))     # here we can change the  fig size by this  command  (height,width)
plt.pie(gb['Churn'],labels =gb.index, autopct = "%1.2f%%")
plt.title(" PERCENTAGE  OF CUSTOMER  CHURN " , fontsize = 10) 
plt.show()
```

# FROM GIVEN  PIE CHART WE CAN CONCLUDE THAT 26.54 % OF OUR CUSTOMER CHURNED OUT

```python
# LETS  IDENTIY THE CHURN GENDER
plt.figure(figsize =(4,4))
ax=sns.countplot(x='gender' , data =df ,hue ='Churn')
ax.bar_label(ax.containers[0])
plt.title('CHURN BY GENDER')
plt.show
```

```python
# LETS  IDENTIY THE CHURN SENIORCITIZEN
plt.figure(figsize =(4,4) )
ax=sns.countplot(x='SeniorCitizen' , data =df ,hue ='Churn')
ax.bar_label(ax.containers[0])
plt.title('CHURN BY SENIORCITIZEN')
plt.show
```

# COMPARATIVELY A GREATER PERCENTAGE OF PEOPLE IN SENIOR CITIZEN CATEGORY HAVE CHURNED

```python
# LETS  IDENTIY THE CHURN CONTRACT
plt.figure(figsize =(4,4) )
ax=sns.countplot(x='Contract' , data =df ,hue ='Churn')
ax.bar_label(ax.containers[0])
plt.title('CHURN  OF CUSTOMER BY CONTRACT')
plt.show
```

```python


# List of the columns you want to plot
columns_to_plot = ['PhoneService', 'MultipleLines', 'InternetService',
                   'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                   'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set up the matplotlib figure
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 12))  # Adjust size as needed

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop over each column and plot the countplot
for i, col in enumerate(columns_to_plot):
    sns.countplot(x=col, data=df, ax=axes[i], palette="Set2", hue =df['Churn'])  # Choose a color palette
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xticklabels(axes[i].get_xticklabels())#, rotation=45, ha="right")  # Rotate x labels if needed

# Adjust the layout so everything fits without overlap
plt.tight_layout()
plt.show()
```

#The charts depict the relationship between various telecom services and customer churn. Key insights include:
"""
1. Customers with phone services show a lower churn rate, whereas those without internet services generally churn less.
2. Fiber optic internet service has a higher churn rate compared to DSL or no internet service.
3. Features like online security, backups, device protection, and tech support are associated with lower churn when availed.
4. Streaming services (TV and movies) seem to have minimal influence on churn compared to other factors.
"""
